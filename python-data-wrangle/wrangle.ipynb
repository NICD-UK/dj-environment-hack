```python
import pandas as pd
import geojson
```

```python
def data2geojson(df, filename):
    features = []
    insert_features = lambda X: features.append(
            geojson.Feature(geometry=geojson.Point((X["Sensor Centroid Longitude"],
                                                    X["Sensor Centroid Latitude"],
                                                    X["Ground Height Above Sea Level"]
                                                   )),
                            properties=dict(name=X["Sensor Name"],
                                            description=X["Sensor Name"],
                                           value=X["Ground Height Above Sea Level"])))
    df.apply(insert_features, axis=1)
    with open(filename, 'w', encoding='utf8') as fp:
        geojson.dump(geojson.FeatureCollection(features), fp, sort_keys=True, ensure_ascii=False)
```

```python
df_data = pd.read_csv("data/data.csv")
df_data.head()
```

```python
df_sensors = pd.read_csv("data/sensors.csv")
df_sensors.head()
```

```python
list_sensor_names = df_data['Sensor Name'].unique()
len(list_sensor_names)
```

```python
list_sensor_names
```

```python
df_sensors = df_sensors[df_sensors['Sensor Name'].isin(list(list_sensor_names))]
df_sensors.head()
```

```python
data2geojson(df_sensors, "data/sensors.geojson")
```
